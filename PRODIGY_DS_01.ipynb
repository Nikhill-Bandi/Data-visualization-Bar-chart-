```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("population.csv")
```

```python
data.head(266)
```

```python
data.describe
```

```python
countries = ['Austria', 'Iraq', 'Australia', 'France','South Africa']

filter_data = data[data['Country Name'].isin(countries)]

plt.bar (filter_data['Country Name'], filter_data['2022'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2022')
plt.title('Population of selected countries in 2022')
plt.xticks(rotation=45, ha='right')
plt.show()
```
